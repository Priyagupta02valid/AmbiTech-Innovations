```python
import pandas as pd

# Load dataset
df = pd.read_csv('AB_NYC_2019.csv')

# Inspect data
print(df.info())
print(df.describe())
print(df.isnull().sum())

# Handle missing values
df = df.dropna(subset=['name', 'host_name'])  # Remove rows where 'name' or 'host_name' is missing
df['reviews_per_month'].fillna(0, inplace=True)  # Fill missing 'reviews_per_month' with 0

# Data types conversion
df['last_review'] = pd.to_datetime(df['last_review'])

# Remove duplicates
df = df.drop_duplicates()

# Handle outliers (example for 'price')
df = df[df['price'] <= df['price'].quantile(0.99)]

# Save cleaned data
df.to_csv('cleaned_AB_NYC_2019.csv', index=False)
```
